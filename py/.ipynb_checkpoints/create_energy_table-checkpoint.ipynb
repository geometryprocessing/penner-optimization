## Energy Tables

```python
def make_record(C, lambdas, lambdas_target, Th_hat):
    record = {}
    record['C'] = C
    record['lambdas'] = lambdas
    record['lambdas_target'] = lambdas_target
    record['Th_hat'] = Th_hat
    
    return record
```

```python
def symmetric_stretches(lambdas, lambdas_target):
    l = np.exp(lambdas/2)
    l_target = np.exp(lambdas_target/2)
    return 0.5*(l/l_target + l_target/l)
```

```python
def C_py_to_cpp(C):
    C_cpp = Connectivity_cpp()
    C_cpp.n = C.next_he
    C_cpp.prev = C.prev_he
    C_cpp.to = C.to
    C_cpp.f = C.he2f
    C_cpp.h = C.f2he
    C_cpp.out = C.out
    C_cpp.opp = C.opp
    
    return C_cpp
```

## Checkpoint Lambdas Table

```python
#pickle_dir = "../output/comparison/conformal/"
#pickle_dir = "../output/incremental/conformal/"
checkpoints_dir = "../output/checkpoints/optimization/"
pickle_dir = "../output/boundary/conformal/"
pickled_logs = os.listdir(pickle_dir)
pickled_logs = [f for f in pickled_logs if f.endswith(".p")]
conf_records = {}
#energies['max_error'] = []
energies['opt_energies'] = []
energies['opt_stretch_max'] = []
energies['opt_stretch_norm'] = []
m_opt = []

for m in models:
    m_checkpoints_dir = os.path.join(checkpoints_dir, m+'_lambdas')
    
    # Get checkpoint lambda files
    if (os.path.isdir(m_checkpoints_dir)):
        m_checkpoints = os.listdir(m_checkpoints_dir)
        m_checkpoints = {int(f.split('_')[-1]):f for f in m_checkpoints if not f.endswith("_init")}
        n_c = len(m_checkpoints)
    else:
        n_c = 0
    
    # If checkpoint lambdas exist
    if (n_c > 0):
        # Get initial lambdas
        lambdas_init_path = os.path.join(m_checkpoints_dir, 'lambdas_init')
        lambdas_init = np.loadtxt(lambdas_init_path,dtype=float)

        # Compute energies of last checkpoint lambdas
        lambdas_path = os.path.join(m_checkpoints_dir, m_checkpoints[n_c-1])
        lambdas = np.loadtxt(lambdas_path,dtype=float)
        energies['opt_energies'].append(energy(lambdas_init, lambdas))
        stretches = symmetric_stretches(lambdas, lambdas_init)
        energies['opt_stretch_max'].append(np.max(stretches))
        energies['opt_stretch_norm'].append(norm(stretches))
        m_opt.append(m)
    else:
        print("Missing opt lambdas for {}".format(m))
        #energies['opt_energies'].append(-1)
        #energies['opt_stretch_max'].append(-1)
```

```python
pickle_dir = "../output/checkpoints/conformal/"
pickled_logs = os.listdir(pickle_dir)
pickled_logs = [f for f in pickled_logs if f.endswith(".p")]
energies['conf_energies'] = []
energies['conf_stretch_max'] = []
energies['conf_stretch_norm'] = []
for m in models_opt:
    pickle_path = os.path.join(pickle_dir, m+'_out.p')
    try:
        with open(pickle_path, 'rb') as f:
            record = pickle.load(f)
            energies['conf_energies'].append(energy(record['lambdas_target'], record['lambdas']))
            stretches = symmetric_stretches(record['lambdas_target'], record['lambdas'])
            energies['conf_stretch_max'].append(np.max(stretches))
            energies['conf_stretch_norm'].append(norm(stretches))
    except FileNotFoundError:
        print("Missing conf energy for {}".format(m))
        energies['conf_energies'].append(-1)
```

```python
energies_df = pd.DataFrame(energies, index=models_opt)
energies_df = energies_df.loc[:,['conf_energies',
                                 'opt_energies',
                                 'conf_stretch_norm',
                                 'opt_stretch_norm',
                                 'conf_stretch_max',
                                 'opt_stretch_max']]
energies_df = energies_df.sort_values('conf_stretch_max')
energies_df
```

```python
def plot_original(pickle_path):
    with open(pickle_path, 'rb') as f:
        record = pickle.load(f)
    C = record['C']
    l = np.exp(record['lambdas_target'] / 2)
    phi_0 = np.zeros(len(C.out))
    make_delaunay(C, l, phi_0, float_type)
    u,v,is_cut_h = layout(C, l, phi_0,start_h=0,float_type=float)
    uv, F, Vn_to_V, _, _, _, _ = apply_cuts(C, is_cut_h, u, v)
    plot.plot(uv, F, shading={'wireframe': True})

```

```python
def plot_pickle(pickle_path):
    with open(pickle_path, 'rb') as f:
        record = pickle.load(f)
    C = record['C']
    l = np.exp(record['lambdas'] / 2)
    phi_0 = np.zeros(len(C.out))
    make_delaunay(C, l, phi_0, float_type)
    u,v,is_cut_h = layout(C, l, phi_0,start_h=0,float_type=float)
    uv, F, Vn_to_V, _, _, _, _ = apply_cuts(C, is_cut_h, u, v)
    plot.plot(uv, F, shading={'wireframe': True})
```
