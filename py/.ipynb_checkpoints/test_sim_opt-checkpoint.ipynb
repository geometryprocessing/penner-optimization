```python
def plot_angles(C, lambdas_init, direction, a=3, n=50):
    X = np.linspace(0,a,n)
    Y_max_F = []
    Y_energy = []
    for gamma in X:
        # Get energy and distortion measures after the step
        lambdas = lambdas_init + gamma * direction
        F, _ = F_with_jacobian_multiple_del(C, lambdas[proj], False, 0)
        energy_step = energy(lambdas, lambdas_target, p=2)
        max_F_step = np.max(np.abs(F))
        max_F_step = norm(F)
        print("Energy at {}: {}".format(gamma, energy_step))
        print("Angle error at {}: {}".format(gamma, max_F_step))
        Y_max_F.append(max_F_step)
        Y_energy.append(energy_step)
    plt.plot(X, Y_max_F)
    print(np.max(Y_energy))
        
    return X, Y_max_F, Y_energy
```

Instead of zigzagging between conformal and optimization methods, we get the sum of these two directions and use a larger multiple of it to extrapolate the flow. Two options for this are (a) to use a full conformal projection and (b) to use a single conformal step. The second option is more local in nature and thus perhaps more reasonable and amenable to rigorous analysis.

```python
def fast_optimize_lambdas(C,
                          lambdas_init,
                          lambdas_target,
                          float_type=float,
                          gamma_0=1.0,
                          checkpoint_dir=None,
                          proj_params=ProjectionParameters(),
                          opt_params={}):
    """
    Use projected gradient descent with learning rate gamma to minimize the l_2 distance of
    the log edge lengths from the initial log edge lenghts subject to the angle constraints
    defined by C.Th_hat.
    """
    log = []
    start = time.time()

    # Build edge and projection maps to go between different representations for the log
    # edge lengths (i.e. embedded mesh edge lengths, full mesh edge lenghts, and full 
    # mesh halfedge lengths)
    he2e, e2he = build_edge_maps(C)
    proj, embed = build_refl_proj(C)
    he2e = np.array(he2e)
    e2he = np.array(e2he)
    proj = np.array(proj)
    embed = np.array(embed)

    # Get parameters for optimization
    num_iter = opt_params['num_iter'] if 'num_iter' in opt_params else 50
    beta_0 = opt_params['beta_0'] if 'beta_0' in opt_params else 1.0
    min_ratio = opt_params['min_ratio'] if 'min_ratio' in opt_params else 1e-2
    require_energy_decr = opt_params['require_energy_decr'] if 'require_energy_decr' in opt_params else True
    max_angle_incr = opt_params['max_angle_incr'] if 'max_angle_incr' in opt_params else math.inf
    max_angle = opt_params['max_angle'] if 'max_angle' in opt_params else math.inf
    num_extra_del = opt_params['num_extra_del'] if 'num_extra_del' in opt_params else 0
    p = opt_params['p'] if 'p' in opt_params else 2


    # Print initial convergence information
    F_init, _ = F_with_jacobian_multiple_del(C, lambdas_init[proj], False, num_extra_del)
    print("Starting optimization", flush=True)
    print("Initial energy: {}".format(energy(lambdas_init, lambdas_target, p=p)))
    print("Initial max angle error: {}\n".format(np.max(np.abs(F_init))))

    lambdas = lambdas_init
    for k in np.arange(num_iter):
        log_k = {}

        # Print start of iteration information
        print("Iteration {}".format(k))
        F_k_init, _ = F_with_jacobian_multiple_del(C, lambdas[proj], False, num_extra_del)
        print("Energy at start of iteration {}: {}".format(k, energy(lambdas, lambdas_target, p=p)))
        print("Max angle error at start of iteration {}: {}\n".format(k, np.max(np.abs(F_k_init))))
        
        # Project to the constraint for the initial lengths
        print("Getting conformal projection direction", flush=True)
        lambdas_proj, u = project_to_constraint(C,
                                                lambdas,
                                                proj_params)
        lambdas_proj = np.array(lambdas_proj)
        conf_dir = lambdas_proj - lambdas
        
        # Print projection information
        print("\nProjection complete", flush=True)
        F_proj, _ = F_with_jacobian_multiple_del(C, lambdas_proj[proj], False, num_extra_del)
        print("Energy in conformal direction: {}".format(energy(lambdas_proj, lambdas_target, p=p)), flush=True)
        print("Max angle error in conformal direction: {}\n".format(np.max(np.abs(F_proj))))

        # Get descent direction
        print("Getting optimized direction", flush=True)
        start_call = time.time()
        F_k, J_F_k = F_with_jacobian_multiple_del(C,
                                                  lambdas_proj[proj],
                                                  True,
                                                  num_extra_del)

        # Get descent direction from constraint function and Jacobian
        delta_lambdas_k = line_search_direction(C,
                                                lambdas_target,
                                                lambdas_proj,
                                                F_k,
                                                J_F_k)
        delta_lambdas_k = np.array(delta_lambdas_k)
        delta_lambdas_norm_k = norm(delta_lambdas_k)
        log_k['delta_lambdas_norm'] = delta_lambdas_norm_k
        print('Finding direction took {} s'.format(time.time() - start_call), flush=True)
        print('Delta lambdas norm {} in iteration {}'.format(delta_lambdas_norm_k, k))

        # Get optimized descent direction
        fast_opt_dir_k = delta_lambdas_k + conf_dir
        print('Fast optimization direction norm {} in iteration {}\n'.format(norm(fast_opt_dir_k), k))
        
        # Reduce line step until angles decrease
        gamma_k = gamma_0
        while True:
            lambdas_k = lambdas + gamma_k * (gamma_k * delta_lambdas_k + conf_dir)

            # Shrink gamma until angles decrease
            F_k, _ = F_with_jacobian_multiple_del(C, lambdas_k[proj], False, num_extra_del)
            print(np.max(np.abs(F_k_init)))
            print("Max angle after step with gamma {}: {}".format(gamma_k, np.max(np.abs(F_k))))
            if gamma_k < 1e-10:
                break
            elif (np.max(np.abs(F_k)) > np.max(np.abs(F_k_init))):
                gamma_k /= 2
            else:
                break
        # Get energy and distortion measures after the step     
        energy_step_k = energy(lambdas_k, lambdas_target, p=p)
        energy_full_step_k = energy(lambdas_k[proj], lambdas_target[proj], p=p)
        max_F_step_k = np.max(np.abs(F_k))
        max_stretch_step_k = np.max(symmetric_stretches(lambdas_k, lambdas_target))
        max_stretch_prev_step_k = np.max(symmetric_stretches(lambdas_k, lambdas))
        max_stretch_init_step_k = np.max(symmetric_stretches(lambdas_k, lambdas_init))
        time_k = time.time() - start
        log_k['time'] = time_k
        log_k['energy_step'] = energy_step_k
        log_k['energy_full'] = energy_full_step_k
        log_k['max_F_step'] = max_F_step_k
        log_k['max_stretch_step'] = max_stretch_step_k
        log_k['max_stretch_prev_step'] = max_stretch_prev_step_k
        log_k['max_stretch_init_step'] = max_stretch_init_step_k
        print("Total optimization time {}".format(time_k)) 
        print("Energy after line step: {}".format(energy_step_k), flush=True)
        print("Full mesh energy at end of iteration {}: {}".format(k, energy_full_step_k))
        print("Max angle error: {}".format(max_F_step_k)) 
        print("Max stretch relative to target: {}".format(max_stretch_step_k))
        print("Max stretch relative to initial: {}".format(max_stretch_init_step_k))
        print("Max stretch relative to previous: {}\n".format(max_stretch_prev_step_k))

        # Update lambdas
        lambdas = lambdas_k

        # Optionally save the lambdas to file
        if checkpoint_dir:
            checkpoint_file_k = os.path.join(checkpoint_dir, 'lambdas')
            np.savetxt(checkpoint_file_k, lambdas)
    
        # Return if optimization step failed
        if gamma_k < 1e-10:
            break
            
    return log, lambdas
```

```python
proj_params = ProjectionParameters()
proj_params.max_itr = 2
proj_params.do_reduction = True
opt_params = { 'p': 2,
               'require_energy_decr': True }
opt_params['min_ratio'] = 1e-5
opt_params['num_iter'] = 50
opt_params['max_angle'] = 10

log_fast_opt, lambdas_fast_opt = fast_optimize_lambdas(C,
                                                       lambdas_init,
                                                       lambdas_target,
                                                       gamma_0=1.0,
                                                       proj_params=proj_params,
                                                       opt_params=opt_params)
```
