```python
# Get the mesh file location
data_dir = "../data/all_scale_diff_patches/"
data_dir = "../data/extreme_scale_diff_patches_2"
data_dir = "../data/for_symm_lengths"
#data_dir = "../data/for_symm_lengths_no_reroute_all"
#data_dir = "../data/fixed_extra_cut"
#data_dir = "../data/fixed_extra_cut_additional"
#data_dir = "../output/gearbox_optimize"
suffix = ''
files = os.listdir(data_dir)
models = [f[:-(4+len(suffix))] for f in files if f.endswith(suffix+".obj")]

# Get complete output files
#output_dir = '../output/hanxiao_meshes'
output_dir = "../output/sim_opt_extreme_2_small"
output_dir = "../output/for_symm_lengths"
output_dir = "../output/for_symm_lengths_continue"
#output_dir = "../output/for_symm_lengths_no_reroute"
#output_dir = "../output/fixed_extra_cut"
#output_dir = "../data/fixed_extra_cut_additional"
#output_dir = "../output/gearbox_optimize"
output_files = os.listdir(output_dir)
finished_models = [f[:-9] for f in output_files if f.endswith("_record.p")]
unfinished_models = np.setdiff1d(models, finished_models)

print("Unfinished models")
unfinished_models
```

```python
# Initialize mesh from the data set for projection and he2e
for m in models:
    print(m)
    v, f = igl.read_triangle_mesh(data_dir+'/'+m+suffix+'.obj')
    Th_hat = np.loadtxt(data_dir+"/"+m+'_Th_hat',dtype=float)
    C, vtx_reindex = fv_to_double(v, f, Th_hat, False)
    proj, embed = build_refl_proj(C)
    he2e, e2he = build_edge_maps(C)
    proj = np.array(proj)
    he2e = np.array(he2e)

    # Get optimized lambdas from file
    lambdas_path = os.path.join(output_dir, m + '_lambdas', 'lambdas')
    lambdas_target = lambdas_from_mesh(C)
    lambdas = np.loadtxt(lambdas_path,dtype=float)
    while (len(lambdas) != len(lambdas_target)):
        lambdas = np.loadtxt(lambdas_path,dtype=float)
        print("Reloading")

    # Check angle error
    F, _ = F_with_jacobian(C, lambdas[proj], False)
    print("Max F: {}\n".format(np.max(np.abs(F))))

    # Save he length array to file
    os.makedirs(output_dir+'_l',exist_ok=True)
    l_path = os.path.join(output_dir+'_l', m + '_l')
    np.savetxt(l_path, np.exp(lambdas[proj[he2e]] / 2)) 
```
